```python
dataset = pd.read_csv(f'201908_new.csv')
dataset['utc_time'] = pd.to_datetime(dataset['utc_time'])
    # Split into separate date and time columns
dataset['date'] = dataset['utc_time'].dt.date
dataset['time'] = dataset['utc_time'].dt.time
dataset['time_numeric'] = dataset['time'].apply(lambda x: x.hour + x.minute)
dataset.sort_values(by= ['date', 'time'], inplace = True)
dataset
```

“max”: giá trị điểm ảnh lớn nhất trong một ảnh.

```python
# Sort the DataFrame by 'value' column before grouping
sorted_dataset = dataset.sort_values(by='value')

# Group by 'date' and 'time_numeric' and get the maximum value for each group
values_pixel_per_hours_max = sorted_dataset.groupby(['date', 'time_numeric'])['value'].max().sort_values()

# Now you can proceed with visualizing or analyzing the 'values_pixel_per_hours_max' data
values_pixel_per_hours_max.sort_values()
```

Lượng mưa lớn nhất tại một điểm ảnh: 88.9mm


```python
count = 0
for i in range(80):
    if values_pixel_per_hours_max[i] > 20:
        count = count + 1

print(count)
```

Số lượng ảnh không có điểm ảnh vượt ngưỡng 20 mm: 9

Số lượng ảnh có điểm ảnh vượt ngưỡng 20mm: 71

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = values_pixel_per_hours_max.values
plt.figure(figsize = (10, 5))

plt.plot(ypoints)
max_rainfall = 88.9
plt.scatter(ypoints.argmax(), max_rainfall, color = '#88c999',  label=f'max')
plt.text(ypoints.argmax(), max_rainfall,'88.9', ha = 'right')

plt.axhline(y=20, linestyle='--')
# Find the x-coordinate (index) where the line crosses the 20mm dashed line
x_intersection = np.argwhere(np.diff(np.sign(ypoints - 20))).flatten()

# Add a point at the intersection
if len(x_intersection) > 0:
    plt.scatter(x_intersection, [ypoints[x_intersection]], color='orange', label='điểm ngưỡng')


plt.legend()
plt.xlabel('Ảnh')
plt.ylabel('Rainfall (mm)')
plt.title('Giá trị điểm ảnh lớn nhất trong mỗi ảnh/ Lượng mưa lớn nhất tại mỗi điểm ảnh')



plt.show()
```
